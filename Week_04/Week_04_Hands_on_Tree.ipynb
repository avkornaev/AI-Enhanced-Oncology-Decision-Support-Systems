```python
import matplotlib.pyplot as plt
from sklearn import tree
import numpy as np

# -----------------------------
# Improved toy dataset: fruits
# Now neither feature alone suffices
# -----------------------------
# Features: [diameter (cm), color (0=green, 1=red, 2=orange)]
X = np.array([
    [4, 0],  # small green -> lime
    [5, 0],  # small green -> lime
    [7, 1],  # medium red -> apple
    [8, 1],  # large red -> apple? Wait...
    [6, 2],  # medium orange -> orange
    [9, 2],  # large orange -> orange
])

y = ["lime", "lime", "apple", "apple", "orange", "orange"]

feature_names = ["diameter (cm)", "color"]
class_names = ["apple", "lime", "orange"]  # sorted order matters for display

# Train decision tree
clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# Plot tree
plt.figure(figsize=(10, 6))
tree.plot_tree(
    clf,
    feature_names=feature_names,
    class_names=sorted(class_names),  # ensure matching order
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree for Fruit Classification", fontsize=14)
plt.tight_layout()
plt.savefig("improved-fruit-tree.png", dpi=300)
plt.show()
```
